```python
import requests
import ast
# To set up the api keys
from google.colab import userdata

# retrieve api key for geolocation
ABSTRACT_API_KEY = userdata.get('ABSTRACT_API_KEY')

# retrieve api key for ebird
EBIRD_API_KEY = userdata.get('EBIRD_API_KEY')

# check if the api keys were actually retrieved
if EBIRD_API_KEY:
  print("Ebird successful")
else:
  print("ebird failed")


if ABSTRACT_API_KEY:
  print("abstract successful")
else:
  print("abstract failed")


```

```python
# Get the location of the user using their IP address
def get_location():
    url = f"https://ipgeolocation.abstractapi.com/v1/?api_key={ABSTRACT_API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data['city'], data['region'], data['country_code']  # Return city, state/region, and country code
    else:
        print("Failed to get location")
        return None, None, None
```

```python
def birds_in_location(country_code):
    # Convert the region to a code if necessary (e.g., using a mapping)
    # country_code = f"{region.lower()}-{city.lower()}"  # Example format, adjust as necessary
    url = f"https://api.ebird.org/v2/data/obs/{country_code}/recent"
    headers = {
        'X-eBirdApiToken': EBIRD_API_KEY
    }

    # Make the GET request to the eBird API
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        raw_data = response.json()
        bird_data = {}
        for index in range(len(raw_data)):
          bird_data[raw_data[index]['comName']] = [raw_data[index]['locName'], raw_data[index]['obsDt']]
        return bird_data
    else:
      print("Failed to return bird sightings")
      return []
```

```python
def main():
    print("Fetching your location...")
    city, region, country_code = get_location()
    if city and region and country_code:
        print(f"The city you are in is: {city}, {region}, {country_code}")
        sightings = birds_in_location(country_code)
        if sightings:
            print("Recent bird sightings are as follows")
            print("Name: Location, Date-Time")
            for name,values in sightings.items():
              print(f"{name}: {values[0]}, {values[1]}")
        else:
            print("No recent bird sightings found.")
    else:
        print("Unable to fetch location.")
if __name__ == "__main__":
    main()
```
